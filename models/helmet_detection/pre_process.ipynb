L'objectif de ce notebook est de séparer les données en train, dev et test sets.

```python
%matplotlib inline
import random
from PIL import Image
import matplotlib.pyplot as plt
```

```python
n_images = 764
```

```python
p_train = 0.95
p_dev = 0.025
p_test = 0.025
n_train = p_train * n_images
n_dev = p_dev * n_images
n_test = p_test * n_images
print("n_train = ", n_train)
print("n_dev = ", n_dev)
print("n_test = ", n_test)
```

```python
random.seed(1)
t = [0] * n_images
for i in range (n_images):
  r = random.random()
  if (r > p_train and r < p_train + p_dev): t[i] = 1
  if (r >= p_train + p_dev ): t[i] = 2

train_set = [i for i in range(n_images) if t[i] == 0]
dev_set = [i for i in range(n_images) if t[i] == 1]
test_set = [i for i in range(n_images) if t[i] == 2]
n_train_real, n_dev_real, n_test_real = len(train_set), len(dev_set), len(test_set)
```

```python
print("Train set de ", n_train_real, "images : ", train_set)
print("Dev set de ", n_dev_real, "images : ", dev_set)
print("Test set de ", n_test_real, "images : ", test_set)
```

## Affichage des tailles des images

```python
filename = "/content/drive/MyDrive/Hackathon IA/Bike helmet detection/Images/BikesHelmets"
width = []
height = []
for i in range (n_images):
  image = Image.open(filename + str(i) + ".png")
  w, h = image.size
  width.append(w)
  height.append(h)
```

```python
print("Min width = ", min(width))
print("Average width = ", sum(width) / n_images)
print("Max width = ", max(width))
print("Min height = ", min(height))
print("Average width = ", sum(height) / n_images)
print("Max height = ", max(height))
plt.hist(width, density=True, bins=30)  # density=False would make counts
plt.ylabel('Distribution width')
plt.xlabel('Width')
plt.show()
plt.hist(height, density=True, bins=30)  # density=False would make counts
plt.ylabel('Distribution height')
plt.xlabel('Height')
plt.show()
```
